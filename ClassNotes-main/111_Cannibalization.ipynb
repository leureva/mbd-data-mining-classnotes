# Cannibalization

Any change creates two types of results: those that are intended, and those that are unintended. 

> In marketing strategy, `cannibalization` refers to a reduction in sales volume, sales revenue, or market share of one product as a result of the introduction of a new product by the same producer. 

Cannibalization is directly linked to anything that changes the products on sale. 

 
By making the change, we are looking to achieve a specific outcome: 
* selling more item 
* making more money selling the items we already sell. 

Typically, a new product adversely affects the sales of existing products: 
* If it affects your competetior's products, it's `Conquest`
* If it affects your own products, it's `Cannibalization`

Any change creates two types of results: 
1. those that are intended
    * Cannibalization - the reduction is a direct result of the new product.    
      * other factors, may be at play, that could cause the reduction
      * so, we have to examine the individual transactions with the new product on it. 
    * Cannibalization cant be measured  by simply looking at a products combinations, because it does not shows the content of each transactions.

2. those that are unintended. 
 


Market Basket is a term for describing all the items that are sold in combination with another item.  




## Ways to measure cannibalization

Many companies have no way to measure cannibalization in a reasonably accurate manner. Here are three ways to do it right:

1. New product

  * Look at the top five items being sold in combination with the new product. 
  * Then look at the market baskets of those five items before and after the product launch to measure how each of those market baskets have changed. 
  * Were the changes better for you overall, or worse?

2. Price change

  * Measure market baskets before and after the price change to measure the shift in the basket of goods sold with the item whose price is increased. 
  * Measure the bill average of the bills containing the price-changed item. 
  * If you have your item's price, you should do this analysis on a contribution margin basis instead of just a revenue basis.

3. Limited time offer

  * Measure average bill for bills containing the `Limited time offer` items before and after the promotion. 
  * Analyze the market basket associated with the `Limited time offer` items to measure trade ups and trade downs, or shifts to other items.

4. Trade-ins 

  * New and old product not comparable. 

5. Shopers surveys (2nd choice) 

  * Respondents tend to exagerage their conterfactual choice 

5. Choice experiments
  * Hypothetical choices are non-committal
  * Expensive to conduct

## Conclusion

Every change you make creates intended consequences and unintended consequences. 
Simple product mix can tell you how much of an item you sell, but it cannot help you measure shifts in the market basket of goods that are bought in connection with a menu item. Check level analysis is required to accurately measure cannibalization.
