## 1. Accessing first level

````python
# To get all rows
df.loc[:, 'Day']
````

````python
df.loc['London' , 'Day']
````

````python
# To get all columns
df.loc['London' , :]
````

````python
# This also works
df.loc['London' , '01/07/19'].to_frame().transpose()
````

````python
# Equivalent to df.loc['London' , '2019-07-02']
df.loc[('London' , '02/07/19')].to_frame().transpose()
````

````python
# Equivalent to df.loc['London', 'Day']
df.loc[('London' ) , ('Day')]
````

````python
# Equivalent to df.loc[:, 'Day']
df.loc[:, ('Day',)]
````

## 2. Multi-level index

````python
df.loc['London', 'Day'].loc['02/07/19'].to_frame().transpose()
````

````python
# Use tuple
df.loc[('London', '02/07/19'), 'Day'].to_frame().transpose()
````

````python
# Select multiple rows
df.loc[ 
    ('London' , ['01/07/19','02/07/19'] ) ,
    'Day'
]
````

````python
df.columns
````

````python
# Select multiple columns
df.loc[ 
    'London' ,
    ('Day', ['Weather', 'Wind'])
]
````

## 3. Selecting a range of data viaslice

````python
# First level index
df.loc[
    'Cambridge':'Oxford',
    'Day'
]

# Above is a shorthand to
# df.loc[
#     ('Cambridge', ):('Oxford', ),
#     'Day'
# ]
````

````python
# We are getting a SyntaxError
#df.loc[
#    ('London', '2019-07-01':'2019-07-03'),
#    'Day'
#]
````

````python
# Correct way to slice
df.loc[
    ('London', '01/07/19'):('London','03/07/19'),
    'Day'
]
````

````python
df.loc[
    ('Cambridge', '02/07/19'):('London','02/07/19'),
    'Day'
]
````

## 4. Selecting all content using `slice(None)`

Getting SyntaxError
```IPython
df.loc[ 
    ( : , '02/07/19'),
    'Day'
]
```

Getting SyntaxError
```
df.loc[ 
    ('London', :), #works if you remove :
    'Day'
]
```

The correct way to select all content: 

````python
df.loc[ 
    ('London', slice(None)),
    'Day'
]
````

````python
# The correct way to select all content
df.loc[ 
    ('London' ),
    'Day'
]
````

## Only of a particular date

````python
# The correct way to select all content
df.loc[
    (slice(None) , '02/07/19'), 
    'Day'
]
````

## 5. Using cross-section `xs()`

````python
# The xs() method of DataFrame additionally takes a level argument 
# to make selecting data at a particular level of a MultiIndex easier.
df.xs('02/07/19', 
      level='Date'
     )
````

````python
# multiple keys
df.xs(
    ('London', '02/07/19'), 
      level=['City','Date']
     )
````

````python
# You can also select on the columns with xs, by providing the axis argument.
df.xs('Wind', 
      level=1, 
      axis=1
     )
````

````python
df.xs('Wind', 
      level=1, 
      axis=1, 
      drop_level=False
     )
````
