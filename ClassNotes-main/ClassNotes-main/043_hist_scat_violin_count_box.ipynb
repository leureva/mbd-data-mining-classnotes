Compute the x and y coordinates for points on a sine curve: 

```python
x = np.arange(0, 3 * np.pi, 0.1) # numpy.arange(start, stop, step)
x
```

```python
y = np.sin(x)
y
```

To plot the points using matplotlib: 

```python
plt.plot(x, y)

plt.show()  # You must call plt.show() to make graphics appear.
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)


plt.title('Sine and Cosine',fontsize=17)
plt.legend(['Sine', 'Cosine'],fontsize=17)

plt.xlabel("X",fontsize=17)
plt.ylabel('y',fontsize=17)

plt.yticks(fontsize=17)
plt.xticks(fontsize=17)

plt.show()
```

### Subplots

```python
#plt.rcParams["figure.figsize"] = (10,8)
plt.subplot(2, 1, 1) #subplot grid that has 2 rows and 1 column, position 1

plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)#active 2nd position
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```

## Seaborn

Seaborn is complimentary to Matplotlib and it specifically targets statistical data visualization. But it goes even further than that: Seaborn extends Matplotlib and makes generating visualizations convenient.

While Matplotlib is a robust solution for various problems, Seaborn utilizes more concise paramesters for ease-of-use.

### Scatterplots with seaborn



```python
import seaborn as sns

path = "datasets/Cartwheeldata.csv"

df = pd.read_csv(path)
 
sns.lmplot(x='Wingspan', 
           y='CWDistance', 
           data=df
          )
plt.title('Scatterplot Wingspan vs CWDistance')
plt.show()
```

## Scatter Plots with ``plt.scatter``

A more powerful method of creating scatter plots is the ``plt.scatter`` function, which can be used very similarly to the ``plt.plot`` function:

```python
plt.scatter(x, y, 
            marker='o');
```

The major difference of ``plt.scatter`` from ``plt.plot``:  
- it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

Let's show this by creating a scatter plot with points of many colors and sizes.

In order to better see the overlapping results, we'll also use the ``alpha`` keyword to adjust the transparency level:

```python
xy_size = y.size
xy_size
```

```python
rng = np.random.RandomState(0)

colors = rng.rand(xy_size)
sizes = 1000 * rng.rand(xy_size)

plt.scatter(x, y, 
            c=colors, #random  xy_size colors
            s=sizes,  #random  xy_size sizes
            alpha=0.3,
            cmap='Reds'
           )
plt.colorbar();  # show color scale
```

- the color argument is automatically mapped to a color scale 
    - shown here by the ``colorbar()`` command
- the size argument is given in pixels

In this way, the color and size of points can be used to convey information in the visualization, in order to visualize multidimensional data.

For example, we might use the Iris data from Scikit-Learn, where each sample is one of three types of flowers that has had the size of its petals and sepals carefully measured:

### ``plot`` Versus ``scatter`` which is better? 

``plt.plot`` can be noticeably more efficient than ``plt.scatter``

-  ``plt.scatter`` has the capability to render a different size and/or color for each point, so the renderer must do the extra work of constructing each point individually.
- in ``plt.plot`` the points are always essentially clones of each other, so the work of determining the appearance of the points is done only once for the entire set of data.

For large datasets, the difference between these two can lead to vastly different performance, and for this reason, ``plt.plot`` should be preferred over ``plt.scatter`` for large datasets.

## Histogram ( Distribution Plot )

```python
my_data =  df.CWDistance
my_data
```

```python
sns.histplot(my_data);
```

```python
plt.hist(my_data);
```

```python
plt.hist(my_data, 
         bins = 10,  #default bins are 10
         density = True,  #Standardized, y value goes to 0 - 1 
         alpha = 0.9, # color density
         histtype='stepfilled', 
         color='steelblue',
         edgecolor='red'
        );
```

If you would like to simply compute the histogram (that is, count the number of points in a given bin) and not display it, the ``np.histogram()`` function is available:

```python
counts, bin_edges = np.histogram(my_data, bins=10)
print('counts\t:', counts)
print('edges\t:', bin_edges)
```

```python
plt.hist(my_data, 
         bins = 10, #or a list 
         density = True,  #Standardized, y value goes to 0 - 1 
         alpha = 0.9, # color density
         histtype='stepfilled', 
         color='steelblue',
         edgecolor='red', 
        )
plt.xticks(bin_edges);
```

```python
count, bin_edges = np.histogram(my_data)

my_data.plot(kind='hist', 
                figsize=(10, 5), 
                xticks=bin_edges
               )

plt.title('Histogram of ....')  
plt.ylabel('Count of ....')  
plt.xlabel('Bins and edges of ...')  

plt.show()
```
